```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('../../data/Sleep_Efficiency.csv')

# If alcohol > 0, then alcohol = 1
df['Alcohol consumption'] = (df['Alcohol consumption'] > 0).astype(int)

# Split the data into two groups
df1 = df[df['Alcohol consumption'] == 1]
df0 = df[df['Alcohol consumption'] == 0]

df1.head()
```

```python
fig = plt.figure(figsize =(10, 7))
 
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])

# Creating plot
bp = ax.boxplot([df0['Sleep efficiency'], df1['Sleep efficiency']])
 
# show plot
plt.show()
```
